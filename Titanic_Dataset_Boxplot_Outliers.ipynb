```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load the Titanic dataset
titanic_df = pd.read_csv('titanic-1.csv')
#Create boxplots to visualize outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic_df[['Age', 'Fare']], orient='v')
plt.title('Boxplot of Age and Fare')
plt.ylabel('Value')
plt.xlabel('Feature')
plt.xticks(ticks=[0, 1], labels=['Age', 'Fare'])
plt.show()

#Identify outliers based on boxplot visualization
#visually identify the points that lie outside the whiskers of the boxplot.
#Remove outliers from the dataset
#Remove outliers from the 'Age' column only

# Calculate the first and third quartiles
Q1 = titanic_df['Age'].quantile(0.25)
Q3 = titanic_df['Age'].quantile(0.75)
# Interquartile range (IQR)
IQR = Q3 - Q1
# Calculate lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
# Remove outliers from the dataset
titanic_df = titanic_df[(titanic_df['Age'] >= lower_bound) & (titanic_df['Age'] <= upper_bound)]

#Print the cleaned dataset
print("Cleaned Dataset:")
print(titanic_df)
```
