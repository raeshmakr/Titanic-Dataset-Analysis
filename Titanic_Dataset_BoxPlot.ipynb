```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Titanic dataset
titanic_df = pd.read_csv('titanic-1.csv')
# Identify numerical columns (Age and Fare)
numerical_columns = ['Age', 'Fare']
# Create boxplots to visualize outliers
plt.figure(figsize=(12, 6))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(1, 2, i)
    sns.boxplot(data=titanic_df[column], orient='v')
    plt.title('Boxplot of {}'.format(column))
    plt.ylabel(column)

plt.tight_layout()
plt.show()
```
